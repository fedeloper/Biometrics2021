## Visualize Data

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

Single image

```python
train_dataT = train_data.T
plt.imshow(np.array(train_dataT.loc["Image"][0].split(), dtype=np.uint8).reshape(96,96), cmap='gray')
plt.grid(b=True) #must be set every time the grid must be shown
plt.show()
```

```python
imageData = [[]]*len(train_dataT.loc["Image"])

for ix in range(len(train_dataT.loc["Image"])):
    imgIX = np.array(train_dataT.loc["Image"][ix].split(),dtype=np.uint8).reshape((96,96))
    imageData[ix] = imgIX
```

```python
fig = plt.figure(figsize=(4., 4.)) #might be too small, adjust as needed

rows = 5
cols = 5
nImgs = rows*cols

grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(rows, cols),  # creates 2x2 grid of axes
                 axes_pad=0.01,  # pad between axes in inch.
                 )
for ax, im in zip(grid, imageData[0:nImgs]):
    # Iterating over the grid returns the Axes.
    ax.imshow(im, cmap='gray')
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)

plt.show()
```
